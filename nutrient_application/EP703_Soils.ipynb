```python
# Import database
d=gu.ReadExcel(r'C:\Users\rhember\Documents\Data\EP703\Soils From Jeff Stone\From SOILREGDAT file.xlsx')
```

```python
# Define derived variables
d['Have']=(d['HTOP']+d['HBOT'])/2

# List of organic soil horizons
FloorH=['L','F','H','LF','LH','FH','FL','HF','HL','LFH']
```

```python
d.keys()
```

```python
# Indices to candidate plots (not released, not thinned)
idx={}
# Organic horizons
idx['C BD OH']=np.where( (d['Released']!='Released') & (d['TT']==0) & (d['FT']==0) & (np.isin(d['HORDER'],FloorH)==True) & (d['Have']>=0) & (d['BULK_D_1']>0) )[0]
idx['F BD OH']=np.where( (d['Released']!='Released') & (d['TT']==0) & (d['FT']>0) & (d['FT']<=675) & (np.isin(d['HORDER'],FloorH)==True) & (d['Have']>=0) & (d['BULK_D_1']>0) )[0]
idx['C PC OH']=np.where( (d['Released']!='Released') & (d['TT']==0) & (d['FT']==0) & (np.isin(d['HORDER'],FloorH)==True) & (d['Have']>=0) & (d['C_PER']>0) )[0]
idx['F PC OH']=np.where( (d['Released']!='Released') & (d['TT']==0) & (d['FT']>0) & (d['FT']<=675) & (np.isin(d['HORDER'],FloorH)==True) & (d['Have']>=0) & (d['C_PER']>0) )[0]
# Mineral horizons
idx['C BD MH']=np.where( (d['Released']!='Released') & (d['TT']==0) & (d['FT']==0) & (np.isin(d['HORDER'],FloorH)==False) & (d['Have']>0) & (d['BULK_D_1']>0) )[0]
idx['F BD MH']=np.where( (d['Released']!='Released') & (d['TT']==0) & (d['FT']>0) & (d['FT']<=675) & (np.isin(d['HORDER'],FloorH)==False) & (d['Have']>0) & (d['BULK_D_1']>0) )[0]
idx['C PC MH']=np.where( (d['Released']!='Released') & (d['TT']==0) & (d['FT']==0) & (np.isin(d['HORDER'],FloorH)==False) & (d['Have']>0) & (d['C_PER']>0) )[0]
idx['F PC MH']=np.where( (d['Released']!='Released') & (d['TT']==0) & (d['FT']>0) & (d['FT']<=675) & (np.isin(d['HORDER'],FloorH)==False) & (d['Have']>0) & (d['C_PER']>0) )[0]
```

```python
# Number of unique installations with measurements (Mineral horizons)
print('Number of instalations with control plot BD = ' + str(np.unique(d['INSTAL'][idx['C BD MH']]).size))
print('Number of instalations with fertilized plot BD = ' + str(np.unique(d['INSTAL'][idx['F BD MH']]).size))
print('Number of instalations with control plot C conc = ' + str(np.unique(d['INSTAL'][idx['C PC MH']]).size))
print('Number of instalations with fertilized plot C conc = ' + str(np.unique(d['INSTAL'][idx['F PC MH']]).size))
```

```python
# Number of unique installations with measurements (O horizons)
print('Number of instalations with control plot BD = ' + str(np.unique(d['INSTAL'][idx['C BD OH']]).size))
print('Number of instalations with fertilized plot BD = ' + str(np.unique(d['INSTAL'][idx['F BD OH']]).size))
print('Number of instalations with control plot C conc = ' + str(np.unique(d['INSTAL'][idx['C PC OH']]).size))
print('Number of instalations with fertilized plot C conc = ' + str(np.unique(d['INSTAL'][idx['F PC OH']]).size))
```

# Plot relationship between bulk density (or C concentration) vs. depth for mineral soil horizons

```python
fig,ax=plt.subplots(1,2,figsize=gu.cm2inch(20,9)); 
ms=3
ax[0].plot(d['Have'][idx['C BD MH']],d['BULK_D_1'][idx['C BD MH']],'o',markerfacecolor='None',markeredgecolor=[0.27,0.49,0.79],ms=ms,label='Control plots')
ax[0].plot(d['Have'][idx['F BD MH']],d['BULK_D_1'][idx['F BD MH']],'s',markerfacecolor='None',markeredgecolor=[0.4,0.9,0],ms=ms,label='Fertilized plots')
ax[0].set(ylabel='Bulk density (g/cm3)',xlabel='Depth (cm)')
ax[0].legend(loc='lower right',frameon=False)
ax[0].yaxis.set_ticks_position('both'); ax[0].xaxis.set_ticks_position('both')

ax[1].plot(d['Have'][idx['C PC MH']],d['C_PER'][idx['C PC MH']],'o',markerfacecolor='None',markeredgecolor=[0.27,0.49,0.79],ms=ms,label='Control plots')
ax[1].plot(d['Have'][idx['F PC MH']],d['C_PER'][idx['F PC MH']],'s',markerfacecolor='None',markeredgecolor=[0.4,.9,0],ms=ms,label='Fertilized plots')
ax[1].set(ylabel='Carbon concentration (%)',xlabel='Depth (cm)')
ax[1].yaxis.set_ticks_position('both'); ax[1].xaxis.set_ticks_position('both')
gu.axletters(ax,plt,0.04,0.92)
```

```python
fig,ax=plt.subplots(1,figsize=gu.cm2inch(9,7)); 
ms=3
ax.plot(d['Have'][idx['C PC OH']],d['C_PER'][idx['C PC OH']],'o',markerfacecolor='None',markeredgecolor=[0.27,0.49,0.79],ms=ms,label='Control plots')
ax.plot(d['Have'][idx['F PC OH']],d['C_PER'][idx['F PC OH']],'s',markerfacecolor='None',markeredgecolor=[0.4,.9,0],ms=ms,label='Fertilized plots')
ax.set(ylabel='Carbon concentration (%)',xlabel='Depth (cm)')
ax.yaxis.set_ticks_position('both'); ax.xaxis.set_ticks_position('both')
```
