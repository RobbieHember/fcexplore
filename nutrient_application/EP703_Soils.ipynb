# Assess natural variability within installations

```python
# Get list of unique installation and plots
ind=np.where(np.abs(d['Have']-10)<5)[0]
uI=np.unique(d['INSTAL'][ind])

# Calculate mean absolute difference in C concentration of the upper mineral soil horizon between plots of the same installation
MAD_CC=np.array([])
MADR_CC=np.array([])
for iI in range(uI.size):
    indI=np.where( (d['INSTAL']==uI[iI]) & (np.abs(d['Have']-10)<5) & (d['C_PER']>=0) )[0]
    
    uP=np.unique(d['PLOT'][indI])
    if uP.size<=1:
        continue
    C_PER_mu=np.array([])
    for iP in range(uP.size):
        ind=np.where( (d['INSTAL']==uI[iI]) & (d['PLOT']==uP[iP]) & (np.isin(d['HORDER'],FloorH)==False) & \
                     (np.abs(d['Have']-10)<10) & (d['C_PER']>=0) )[0]
        C_PER_mu=np.append(C_PER_mu,np.mean(d['C_PER'][ind]))
    
    a,b=np.meshgrid(C_PER_mu,C_PER_mu)
    c=a-b
    cr=(a-b)/b*100
    for i in range(c.shape[0]-1):
        for j in range(1,c.shape[1]):
            if i==j:
                continue
            MAD_CC=np.append(MAD_CC,np.abs(c[i,j]))
            MADR_CC=np.append(MADR_CC,np.abs(cr[i,j]))
```

```python
# Calculate mean absolute difference in bulk density of the upper mineral soil horizon between plots of the same installation
MAD_BD=np.array([])
MADR_BD=np.array([])
for iI in range(uI.size):
    indI=np.where( (d['INSTAL']==uI[iI]) & (np.abs(d['Have']-10)<5) & (d['BULK_D_1']>=0) )[0]
    
    uP=np.unique(d['PLOT'][indI])
    if uP.size<=1:
        continue
    #if uP.size==2:
    #    break
    mu=np.array([])
    for iP in range(uP.size):
        ind=np.where( (d['INSTAL']==uI[iI]) & (d['PLOT']==uP[iP]) & (np.isin(d['HORDER'],FloorH)==False) & \
                     (np.abs(d['Have']-10)<10) & (d['BULK_D_1']>=0) )[0]
        mu=np.append(mu,np.mean(d['BULK_D_1'][ind]))
    
    a,b=np.meshgrid(mu,mu)
    c=a-b
    cr=(a-b)/b*100
    for i in range(c.shape[0]-1):
        for j in range(1,c.shape[1]):
            if i==j:
                continue
            MAD_BD=np.append(MAD_BD,np.abs(c[i,j]))
            MADR_BD=np.append(MADR_BD,np.abs(cr[i,j]))
```

```python
fig,ax=plt.subplots(2,2,figsize=gu.cm2inch(20,12)); 
ms=3
ax[0,0].hist(MAD_CC,facecolor=[0.5,0.5,0.5],bins=30)
ax[0,0].set(ylabel='Frequency',xlabel='Mean absolute difference in C concentration (%)')
ax[0,0].yaxis.set_ticks_position('both'); ax[0,0].xaxis.set_ticks_position('both')
txt='Mean = ' + str(np.round(np.mean(MAD_CC),decimals=2)) + '%\nMedian = ' + str(np.round(np.median(MAD_CC),decimals=2)) + '%'
ax[0,0].text(3.5,9,txt)

ax[0,1].hist(MADR_CC,facecolor=[0.5,0.5,0.5],bins=30)
ax[0,1].set(ylabel='Frequency',xlabel='Mean absolute percent difference in C concentration (%)')
ax[0,1].yaxis.set_ticks_position('both'); ax[0,1].xaxis.set_ticks_position('both');
txt='Mean = ' + str(np.round(np.mean(MADR_CC),decimals=2)) + '%\nMedian = ' + str(np.round(np.median(MADR_CC),decimals=2)) + '%'
ax[0,1].text(120,10,txt)

ax[1,0].hist(MAD_BD,facecolor=[0.5,0.5,0.5],bins=30)
ax[1,0].set(ylabel='Frequency',xlabel='Mean absolute difference in bulk density (g/cm3)')
ax[1,0].yaxis.set_ticks_position('both'); ax[1,0].xaxis.set_ticks_position('both')
txt='Mean = ' + str(np.round(np.mean(MAD_BD),decimals=2)) + '%\nMedian = ' + str(np.round(np.median(MAD_BD),decimals=2)) + '%'
ax[1,0].text(0.3,1.7,txt)

ax[1,1].hist(MADR_BD,facecolor=[0.5,0.5,0.5],bins=30)
ax[1,1].set(ylabel='Frequency',xlabel='Mean absolute percent difference in bulk density (%)')
ax[1,1].yaxis.set_ticks_position('both'); ax[1,1].xaxis.set_ticks_position('both');
txt='Mean = ' + str(np.round(np.mean(MADR_BD),decimals=2)) + '%\nMedian = ' + str(np.round(np.median(MADR_BD),decimals=2)) + '%'
ax[1,1].text(35,2.4,txt)
gu.axletters(ax,plt,0.04,0.92)
```

# Quantify the response that exceeds natural variability at specified confidence intervals

```python
# Determine scale parameter that corresponds to 36% error, assuming normal distribution
Scale=45
r=np.random.normal(loc=0,scale=Scale,size=10000)
print(np.mean(np.abs(r)))
```

```python
# Number of instalations
N_Inst=10

# Number of paired plots per installation
N_PlotPairsPerInst=2

# Number of experiments
N_Sim=10000

E_mu=np.zeros(N_Sim)
for i in range(N_Sim):    
    E=np.zeros((N_Inst,N_PlotPairsPerInst))
    for j in range(N_Inst):
        E[j,:]=np.random.normal(loc=0,scale=45,size=N_PlotPairsPerInst)
    E_mu[i]=np.mean(E)

# Calculate percentiles
p95=np.percentile(E_mu,[2.5,97.5])
p50=np.percentile(E_mu,[25,75])
print('95 percent confidence interval = ' + str(p95[0]) + ' to ' + str(p95[1]))
print('50 percent confidence interval = ' + str(p50[0]) + ' to ' + str(p50[1]))
```
