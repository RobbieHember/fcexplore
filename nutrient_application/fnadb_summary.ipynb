## Introduction

This notebook summarizes observed responses to nutrient application in forest ecosystems. The scope includes area-based, multi-year responses of net change in carbon of stemwood, other biomass pools, litter and soils to applications of nitrogen. 

In the case of stemwood, almost all observations represent stemwood volume. Values of stemwood volume were converted to biomass based on the mean wood density of each species. No correction for potential responses of wood density to nutrient applicatoin were applied.

The analysis is based on observations that were compiled in the Forest Nutrient Application Database (FNADB). Due to challenges in acquiring many experiments and the burden of requesting data that are not publically available, and digitizing non-tabular reports, it is not yet quite an exhaustive list of available responses.
